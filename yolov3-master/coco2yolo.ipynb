```python
def img2label_paths(img_paths):
    # Define label paths as a function of image paths
    sa, sb = f'{os.sep}data{os.sep}', f'{os.sep}labels{os.sep}'  # /images/, /labels/ substrings
    return [sb.join(x.rsplit(sa, 1)).rsplit('.', 1)[0] + '.txt' for x in img_paths]
```

```python
import json
import os

# Path to the COCO dataset annotations file
coco_annotations_file = "/Users/shane/Desktop/yolodata/images/annotations.json"

# Path to the COCO dataset images directory
coco_images_dir = "/Users/shane/Desktop/yolodata/images/"

# Path to the YOLO dataset directory
yolo_dataset_dir = "/Users/shane/Desktop/yolodata"

# Create the YOLO dataset directory if it doesn't exist
if not os.path.exists(yolo_dataset_dir):
    os.makedirs(yolo_dataset_dir)

# Load the COCO annotations file
with open(coco_annotations_file, "r") as f:
    annotations = json.load(f)

# Loop over the annotations
for ann in annotations["annotations"]:
    # Get the image ID
    image_id = ann["image_id"]

    # Get the image file name
    image_file_name = annotations["images"][image_id]["file_name"]

    # Get the image width and height
    image_width = annotations["images"][image_id]["width"]
    image_height = annotations["images"][image_id]["height"]

    # Get the bounding box coordinates
    bbox = ann["bbox"]

    # Convert the bounding box coordinates to YOLO format
    x_center = bbox[0] + (bbox[2] / 2)
    y_center = bbox[1] + (bbox[3] / 2)
    width = bbox[2]
    height = bbox[3]

    # Normalize the bounding box coordinates
    x_center /= image_width
    y_center /= image_height
    width /= image_width
    height /= image_height

    # Get the class label
    class_id = ann["category_id"]

    # Create the YOLO annotation file
    yolo_annotation_file_name = os.path.splitext(image_file_name)[0] + ".txt"
    yolo_annotation_file_path = os.path.join(yolo_dataset_dir, yolo_annotation_file_name)
    os.makedirs(os.path.dirname(yolo_annotation_file_path), exist_ok=True)

    # Write the YOLO annotation to file
    with open(yolo_annotation_file_path, "a") as f:
        f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
```
